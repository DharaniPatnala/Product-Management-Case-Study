**Canva – Interpreting Feedback to Shape Product–Market Fit**

As a product manager at Canva, the key responsibility is to continuously improve the product experience to achieve strong product-market fit in terms of user needs and business goals.

 Recently, the feedback from the user survey has been: “the editor is easy to use,” and “too many templates feel generic.” However, at the same time, the business is experiencing a decline in the retention rate of first-time users.

**The feedback from the user survey is an important product insight:**

 “Canva is great at usability and accessibility but poor at delivering personalized and emotionally engaging design experiences.”

 Therefore, in order to improve the product experience and achieve strong product-market fit, the key would be to utilize the product management framework to change the product roadmap at Canva.

 Understanding the Feedback Using Product Frameworks

**Product-Market Fit**

The product-market fit pyramid model is based on the alignment of six factors: the customer segment, underserved needs, value proposition, feature set, user experience, and metrics.

**Target customer segment:** Students, content creators, marketers, educators, entrepreneurs, and working professionals.

**Core user need:** To make design creation easy, fast, and of high quality.

**Key pain point:** Templates lack personalization and differentiation.

Canva meets the user experience requirements but misses the mark in terms of the key pain point. As a result, the level of emotional engagement is low. Therefore, the level of attachment to the product is also low, leading to low retention rates, especially among first-time users.

**Jobs To Be Done (JTBD) Framework**

According to the JTBD model, the users “hire” Canva to get things done in the following ways:

1.   To create impactful social media content.
2.   To design professional resumes.

3.   To create business presentations.
4.   To create marketing materials.

The generic nature of the templates does not closely match the JTBD model. The users want context-aware designs that match their industry, role, and objective.

**AARRR Funnel Analysis**

From the analysis of the AARRR funnel, the following was noted:

**Acquisition** - Strong, given the Canva brand and marketing efforts

**Activation** - High, given the simplicity of the Canva editor

**Retention** - Low, given the lack of emotional engagement from the use of generic templates

This analysis confirms the assumption that the product-market fit issue lies in the area of retention, as opposed to activation or acquisition.

**Duolingo – Prioritizing Features for User Retention**

**Introduction**

As a product manager at Duolingo, the key objective is to increase daily active user (DAU) retention through the development of engaging and habit-forming learning tools. A survey of Duolingo users has revealed two key findings:

    “I love the streak feature – it keeps me coming back.”

    “I don’t find the leaderboard feature motivating – it is irrelevant.”

Furthermore, the RICE analysis has been conducted on five new gamification features to be incorporated into the Duolingo application.

 To create an effective product roadmap, the Kano model and RICE framework would be used to create maximum emotional engagement and business results.

Understanding User Motivation through the Kano Model
The Kano model is used to analyze the features of the product on the basis of their impact on the satisfaction of the end users of the product. There are three types of features in the Kano model: Must-Have, Performance, and Delighters.

Mapping Duolingo Features

The features of the Duolingo application have been mapped to the Kano model as follows:

**Streaks: Performance-Delighters**

Strongly reinforce daily habit formation

Generate Emotional Commitment

**Leaderboards: Indifferent**

Limited Emotional Resonance

Low Impact on Retention

RICE Framework for Data-Driven Prioritization

RICE prioritizes projects based on the following factors:

- **Reach**: The extent to which the project would reach users.
- **Impact**: The effect of the project on the users.
- **Confidence**: The confidence level of the project.
- **Effort**: The effort required to implement the project.

The Streak enhancements scored the highest as the project would reach almost all users, have a high impact on the users, would be highly confident, and would require moderate effort.

**Strategic Decision**: Combining Kano + RICE

**Decision Logic**

**Kano Insight**: Streaks delight users, which means users have an emotional attachment to the product.

**RICE Insight**: Streaks have the highest ROI.

**Conclusion**:
The product development team would focus more on the Streak enhancements, whereas the Leaderboard enhancements would be de-prioritized.

**Product Roadmap Actions**
1. Advanced Streak Enhancements

- Milestone rewards (7, 30, 100-day streaks)
- Personalized motivational reminders
- Visual learning journey timelines
- Streak recovery challenges

2. Smart Behavioral Nudges

    Scheduling reminders using AI

    Adapting lesson difficulty

    Sending personalized motivational messages

This will help in maintaining the engagement without over-engaging the user.

3. Collaborative Gamification (Alternative to Leaderboards)

Instead of the conventional competitive leaderboard system:

    Implementing friend group challenges

    Implementing team-based learning goals

    Implementing community progress milestones

This is more likely to create positive reinforcement and cooperation, which is more in sync with the concept of intrinsic motivation.

**Key Performance Metrics**


*   DAU/MAU ratio
*   Continuation of streaks


*   Day 7 and Day 30 retention
*   Frequency of sessions


**Conclusion**

With the help of the Kano model and the RICE framework, the product decision process is not only emotionally intelligent but also evidence-based. From the analysis, it is evident that the key to retaining users is through the streak-based engagement model, while the leaderboard model has little to offer in the way of motivation.

With the implementation of the streak model, along with the introduction of smart nudges and the collaborative gamification model, Duolingo is likely to see an increase in daily engagement and user loyalty.
