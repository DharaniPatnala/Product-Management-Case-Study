**Designing a Real-Time Telemetry Dashboard for Disney+**

If I were a Product Manager at Disney+ leading this initiative,

I would approach it in three stages:

Understanding user needs,

Defining the right metrics, and

Building a decision-oriented dashboard with clear prioritization.

**1.** **Understanding the Problem & Users**

The core problem is not just lack of data, but lack of actionable visibility. PMs currently rely on fragmented reports, which delays decision-making and creates inconsistencies.

Key Users:

Product Managers (primary users)

Content Strategy Teams
Engineering & Playback Teams

Key Jobs-to-be-Done:

Monitor content performance in real-time

Identify drops in engagement or retention

Detect playback issues quickly

Make faster product/content decisions

Before designing, I would conduct:

Stakeholder interviews

Shadow existing workflows

Identify most frequently used reports

**2.** **Defining Key Metrics** (What to Show)

The dashboard should focus on decision-driving metrics, not vanity metrics.

 I would group them into three layers:

A. User Engagement Metrics

Daily Active Users (DAU) / Monthly Active Users (MAU)
Average Watch Time per User
Session Length
Completion Rate (for shows/movies)
Drop-off points (where users stop watching)

 Insight: Helps PMs understand what users actually watch and enjoy

B. Content Performance Metrics

Top Performing Content (by watch time, views)

Trending Content (real-time spikes)

Content Retention Curve

Genre/category performance

New release performance vs benchmarks

Insight: Helps answer “What content is driving growth?”

C. Playback & Experience Metrics

Video Start Failure Rate

Buffering Frequency & Duration

Playback Error Rate

App Crash Rate

Device/Network-level breakdown

 Insight: Helps identify technical issues impacting experience

**3.** **Dashboard Design Principles** (Ensuring Usability)

To ensure the dashboard is actually used for decision-making, I would focus on:

1. Role-Based Views

PMs see product-level insights

Engineers see deeper technical metrics

Content teams see performance insights

2. Real-Time + Historical Comparison

Show current data alongside trends (e.g., last 7 days)

Highlight anomalies automatically

3. Drill-Down Capability

From high-level metrics → to granular views

Example:

“Drop in watch time → specific show → specific device → specific region”

4. Alerts & Notifications

Automatic alerts for anomalies (e.g., spike in buffering)

Reduces need for constant monitoring

5. Simple, Visual Interface

Use charts, heatmaps, and trend lines

Avoid clutter — prioritize clarity over completeness

**4.** **Prioritization Strategy (What to Build First)**

Given limited resources, I would prioritize based on impact vs urgency:

Phase 1: MVP (High Impact, High Urgency)

Core engagement metrics (DAU, watch time)

Playback health metrics (failures, buffering)

Top content performance

Basic dashboard UI

 Goal: Enable quick visibility into platform health

Phase 2: Insight Layer

Trend analysis & comparisons

Retention curves

Device/region breakdowns

Drill-down capabilities

Goal: Enable root cause analysis

Phase 3: Intelligence Layer

Automated alerts

Predictive insights (e.g., potential churn)

Personalized dashboards

Goal: Enable proactive decision-making

**5.** **Measuring Success of the Dashboard**

To ensure the dashboard delivers value, I would track:

Reduction in time to detect issues

Increase in dashboard adoption by PMs

Faster decision-making cycles

Reduction in playback-related complaints

**Final Thought**

The goal of this dashboard is not just to display data, but to empower faster and better decisions. By focusing on actionable metrics, intuitive design, and phased prioritization, we can transform scattered insights into a centralized, real-time decision engine for Disney+.

**Diagnosing a 20% Drop in Watch Time on New Videos at YouTube**

If I were a Product Manager at YouTube, I would approach this problem systematically in three phases: diagnosis, root cause identification, and solution validation. The key is to move from symptoms (drop in watch time) to actionable insights quickly and accurately.

**1.** **Frame the Problem Clearly**

We are seeing:

20% drop in total watch time

Limited to newly uploaded videos

Upload volume is unchanged

This indicates the issue is likely related to:

Discovery (users not finding videos)

Engagement (users not watching enough)

Experience (technical or UX issues)

**2.** **Diagnose Using Telemetry** (What Signals to Analyze)

I would break down the problem into key metric layers:


A. Discovery Metrics (Are users finding new content?)

Impressions on new videos

Click-Through Rate (CTR) on thumbnails

Traffic sources (Home feed, Search, Recommendations)

 If impressions or CTR dropped → issue in recommendation system or thumbnails


B. Engagement Metrics (Are users watching?)

Average Watch Time per video

Audience Retention Curve (drop-off points)

Watch Time per session

Likes, comments, shares

 If retention drops early → content mismatch or misleading titles/thumbnails


C. Platform/Experience Metrics (Is something broken?)

Video start failure rate

Buffering time & playback errors

App crashes or latency spikes

Device/region-specific issues

If playback issues increased → technical degradation impacting watch time


D. Segmentation Analysis (Where is the drop happening?)

To isolate the issue, I would segment by:

Geography (region-specific issues)

Device type (mobile vs desktop vs TV)

Content category (music, education, entertainment)

User cohort (new vs returning users)

This helps pinpoint whether the issue is localized or systemic

**3.** **Identifying Root Causes** (Hypothesis Building)

Based on the above signals, I would form and test hypotheses such as:

**Hypothesis 1**: Recommendation Algorithm Change

Drop in impressions from Home/Recommended feed
New videos not being surfaced effectively

**Hypothesis 2**: Content Quality / Mismatch

Stable CTR but sharp drop in retention
Indicates users click but leave early

**Hypothesis 3**: Technical Issue

Increase in buffering or playback failures
Particularly in specific regions/devices

**Hypothesis 4**: External Factors

Seasonality or competing events (e.g., major sports or festivals)
Shift in user behavior

**4.** **Deep Dive & Validation**

Once a hypothesis is identified, I would:

Compare before vs after metrics (last 2 weeks vs prior baseline)
Analyze experiment logs or recent releases
Collaborate with:
Engineering (for system changes)
Data Science (for anomaly detection)
Creator teams (for qualitative insights)

**5.** **Solution Approach** (Fixing the Issue)

Depending on the root cause:

If Discovery Issue (Algorithm)
Roll back or tweak ranking model
Boost visibility of new uploads
Rebalance recommendation weights
If Engagement Issue (Content Mismatch)
Improve thumbnail/title guidance for creators
Provide real-time creator analytics
Introduce A/B testing tools for thumbnails
If Technical Issue
Fix playback bugs or latency issues
Optimize video delivery (CDN improvements)
Prioritize high-impact regions/devices

**6.** **Validation of Fix** (Ensuring It Works)

After implementing the fix, I would validate using:

A. A/B Testing
Compare fixed version vs control group
Track watch time, retention, CTR
B. Key Success Metrics
Recovery in total watch time
Improvement in retention curves
Reduction in playback errors
C. Monitoring Dashboards
Real-time alerts for regressions
Continuous tracking over 1–2 weeks

**7.** **Preventing Future Issues**

To ensure this doesn’t recur:

Set up automated anomaly detection alerts
Build real-time telemetry dashboards for early detection
Create post-mortem documentation for learnings

**Final Thought**

The key here is structured thinking: break the problem into discovery, engagement, and experience layers, use telemetry to isolate the issue, and validate solutions through experimentation. At YouTube, this approach ensures we not only fix the drop in watch time but also build a more resilient system for the future.
