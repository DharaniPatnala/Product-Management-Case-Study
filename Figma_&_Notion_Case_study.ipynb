**Structuring Experiments at Figma for"Design-to-Code" Handoff**

**1.** **Problem Context**

At Figma, one of the key areas of friction is observed in the handover from the designer to the developer. Designers at Figma create high-fidelity UI components. However, developers have to interpret these designs on their own. This has resulted in:

Lack of alignment between the designer and the developer
Increased time taken by the developers
Redundancy in the process of clarifying the same thing with both the designer and the developer

To overcome this problem, the team is considering the idea of "Auto Code Suggestions." This is a feature where the designs created by the designer automatically translate into a usable code snippet.

**2.** **Hypothesis**

If Figma is able to provide accurate and usable code snippets from the designs created by the designer, the time taken by the developer would decrease.

**3.** **User Research** (**JTBD + The Mom Test**)

a. Jobs-To-Be-Done (JTBD)

Core Job:

“When I get a design file, I want to be able to quickly turn it into code that’s ready to go into production so that I don’t misinterpret anything.”

Sub-Jobs:

Precisely understanding spacing, typography, styles, etc.
Converting components into code that’s easily reusable
Ensuring consistency with design systems

b. User Segments
Frontend Developers – Target Audience
Product Designers – Secondary Audience
Engineering Managers – Decision Influencers

c. Conducting Interviews (The Mom Test)

Instead of Asking:
❌ “Would you want auto-generated code?”

Ask:

“Tell me about the most recent design from Figma that you implemented.”

“What parts of the design process slow you down the most?”

“How do you currently get code or specs from a design file?”

“Have you used any tools or plugins to help with this process? Did they work well or poorly?”

Key Insights Expected:

Developers rely on inspecting code manually or plugins
Inconsistencies in spacing and styles cause developers to slow down
Trust in auto-generated code is a major pain point

**4.** **Solution Framing**

Auto Code Suggestions Feature:

Translates UI components into framework-specific code (e.g., React, HTML/CSS, etc.)
Guarantees adherence to design systems
Delivers editable, clean, and developer-friendly code snippets

**5.** **Experimentation Strategy**

a. MVP Definition
Initial:

    Limited component types (e.g., buttons, cards, forms)
    One framework (e.g., React)
    Read-only code suggestions

 b. Experiment Phases

Phase 1:

Prototype Testing (Qualitative)

Development of clickable prototypes

Conducting usability tests with 8 to 10 developers

Metrics to be measured:

Ease of understanding code
Perceived usefulness
Trust in output

Phase 2: Beta Launch

Metrics to be measured:

**7.** **Prioritization Using RICE**

For prioritization of the "Auto Code Suggestions"" feature, the RICE framework is used to make sure that a data-driven and systematic approach is adopted for decision-making. The key is to find out which parts of the feature need to be built first to deliver maximum value with minimum risk and development efforts.

From the analysis, Basic UI Component Code Generation, such as buttons, cards, and input fields, is identified as the highest priority. This is because this feature has a high "Reach" and "impact" and a relatively moderate effort.

Conversely, Multi Framework Support, including React, Angular, and Vue, has a high impact but relatively low confidence and effort because of the complexities involved. Therefore, this feature should be considered later after validating this feature.

Editable Code Suggestions is a feature that falls under the medium priority list. It has a relatively high impact but is not a must-have feature for initial feature validation.

**Key Prioritization Insights**

Basic Code Generation for UI Components
High on Reach for Users and Use Cases
High on Tangible Benefits for Developers
Moderate Effort Required
Good ROI for Effort Invested

Defer Multi-Frame Support

High Impact but Difficult to Implement
Less Confidence Due to Differences Across Frameworks
Better to Validate with One First

Editable Code Suggestions to be Added in Phase 2

Improves Flexibility and Control
Not Critical for Initial Validation
Medium Impact and Effort

Defer AI Optimization

High Effort Required
Less Tangible Benefits for Users
Need Strong Foundation for Accurate Code Generation
Can be a Differentiator Later on

**8.** **Risks & Mitigation**

Low Trust in Generated Code
→ Offer transparency (i.e., explain derivation of the code)
Inaccurate Output of Generated Code
→ Begin with basic components, build incrementally
Framework Fragmentation
→ Initial focus on one ecosystem

**9.** **Final Recommendation**

The "Auto Code Suggestions" feature is highly promising in terms of decreasing design-to-development friction. To realize this, we need to ensure that we:

Build trust & accuracy first
Focus on narrow, impactful use cases to begin with
Develop incrementally, not speculatively

**Validating “Auto-Sync Blocks” at Notion**

**1.** **Problem Context**

In Notion, users often have to replicate information such as SOPs, OKRs, policies, and templates on multiple pages. However, this poses a problem in case changes are needed, whereby users have to update each page manually, which results in:

Inconsistent information on different pages
Inefficient update processes
Possibility of outdated or erroneous information being used

To address this problem, the team is considering introducing a feature referred to as "Auto-Sync Blocks," whereby users can update their content on all pages automatically upon a one-time update.

**2.** **Hypothesis**

If users can sync their blocks of content on different pages, they will be able to save time, avoid inconsistencies, and enhance their collaborative work.

**3.** **User Research Strategy**

a. Personas

For relevance, we have identified the following user personas:

Startup Founder / Manager

Needs consistent SOPs and policies

Struggles with maintaining updated documentation

Product / Operations Manager

Manages OKRs, dashboards, and cross-team documentation

Needs real-time updates across multiple pages

Individual Power User

Uses templates and repeated content

Needs efficiency and automation in workflows

b. Use Cases

Updating SOPs across multiple team pages
Syncing OKRs across dashboards and reports
Maintaining consistent company policies
Updating templates used across projects

c. Research Approach

By applying behavioral research and The Mom Test:

Instead of asking:

❌ “Would you use Auto-Sync Blocks?”

Ask:
“Tell me about the last time you updated the same content in multiple places.”
“How do you currently manage repeated content?”
“What happens if one version of the content is outdated?”
“Have you tried any workaround? What didn’t work?”

Expected Insights:

Copy-paste or manual linking of content
High frustration level
High need for a centralized update mechanism

**4.** **Application of Kano Model**

To define feature importance:

Must-Have (Basic Needs)

Reliable syncing (no broken links or errors)

Data consistency across all instances

Performance Needs

Speed of updates across pages
Ease of creating and managing synced blocks

Delighters

Visual indicators of synced content
Partial syncing or overriding of blocks

Insight:

Accuracy and reliability must be prioritized, as any failure in this area will result in a lack of trust from users.

**5.** **Experimentation Strategy**

a. MVP Definition

To be included in the MVP:

Syncing of text blocks only
Manual creation of synced blocks
Visual indicator of synced content

b. Validation Phases

Phase 1: Qualitative Testing

Conduct usability testing with 8 to 12 users

To be evaluated:

Ease of understanding the feature

Trust in syncing behavior

Usefulness of the feature

Phase 2: Beta Rollout

Target a limited user segment
Monitor real-world usage patterns

**6.** **A/B Testing Plan**

Experiment Design

Control Group (A):
Existing Notion experience (manual duplication of content)
Treatment Group (B):
Access to Auto-Sync Blocks
Metrics

Primary Metrics:
% reduction in time spent updating duplicate content
Number of synced blocks created

Secondary Metrics:
% reduction in creating duplicate content
% of users adopting the feature
% of users that are satisfied (CSAT/NPS)

Success Criteria

Achieve at least 30% reduction in content update time
Achieve at least 40% of active users managing multiple pages
Significant reduction in creating duplicate content

**7.** **Prioritization Approach**

Using Structured Prioritization (RICE-Inspired Thinking)

**High Priority**:

Basic sync functionality (value delivery)
Reliability and consistency

**Medium Priority**:

UI indicators for synced blocks
Ability to unlink or edit independently

**Low Priority**:

Advanced customization (partial sync, rules-based sync)
Automation and AI improvements

**8.** **Risks & Mitigation**

User Confusion
→ Provide visual feedback and onboarding
Loss of Control (Overwrites)
→ Provide a way to break the link and override the block
Technical Failures
→ Make sure backend reliability is robust before scaling up

**9.** **Success Metrics Framework**

Adoption Metrics
% Users using the synced blocks
Number of synced blocks created per user
Efficiency Metrics
Time saved per update
Reduction in duplicated content created
Experience Metrics
User satisfaction (CSAT)
Retaining power user base

**10.** **Recommendation**

“Auto-Sync Blocks” has tremendous potential to enhance the documentation process at Notion. Validation should include:

Addressing actual user problems, not theoretical ones
High reliability and trust factors
Start small and build up

By integrating Personas for user clarity, Use Cases for real-world relevance, Kano for feature prioritization, and A/B testing for validation, Notion can ensure that Auto-Sync Blocks delivers meaningful, scalable value rather than just an additional feature.
