Understanding User Needs Through Research: A Case Study of a Student Productivity App
Understanding User Needs Through Research: You’re managing a student productivity app. After a recent update, daily active users have dropped by 30%. How would you use user research to identify and solve the problem? Include how and when you'd apply different types of research methods, and what frameworks or tools you'd use.

**1. Research Objectives**

It is important to clearly define the research objectives before jumping into data collection.
What particular questions must we seek to answer?

Here, my objectives:


*   I would Identify what features or changes exactly in this update have  caused dissatisfaction among users
*  whether it is new bugs, usability issues, or unwanted changes to existing functionality.
*   Are the users shifting to other apps, or is the update confusing or too technical for them?

*   What improvements or fixes would encourage users to return to the app?
*   What features do they miss, what changes would they like to see, and what would make the app more valuable to them?

Segment the user base to understand whether the update has different impacts on different user groups.

**2. Choose Research Methods**

Choosing the appropriate methods is important to get relevant data. This can best be done with a mix method, i.e, quantitative and qualitative methods.


*   Quantitative Data Analysis
*   Qualitative User Research


Quantitative Data Analysis:

APP Analytics: Review app usage data to see if there are specific features that are less utilized since the update. Analyze user behavior for trends like drop-off points in key workflows.

A/B Testing: Whenever possible, revert specific features or changes to their previous state for a small subset of users to analyze whether it impacts their engagement.

Surveys: Run short, focused surveys targeting a large number of end-users to solicit feedback on their overall experience using the update.
Using rating scales and open-ended questions to capture quantitative and qualitative data, respectively.

Qualitative User Research:

Conducting User Interviews: One-on-one interviews are conducted with representative users to probe further into their experiences. Open-ended questions are posed in order for them to talk about motivations, frustrations, and needs.

Usability Testing: Observe users interacting with the updated app in order to identify usability issues and points of confusion. Ask them to think aloud while they complete specific tasks.

Focus Groups: Group small sets of users to discuss their experiences with the update, and come up with possible solutions.
This could be useful in finding pain points that exist across many users, as well as gathering new ideas.

**3. User research**

Having established the research objectives and methods, it's now time to put the plan into action.

Recruiting participants: Carefully choose participants representative of the user base of the application. Consider demographics, patterns of use, and levels of engagement.

Prepare Research Materials: Create interview scripts, usability testing tasks, and survey questionnaires. Make sure the materials are clear, concise, and unbiased.

Conduct Research Sessions: Conduct research sessions with professionalism and empathy. Ensure comfort and safety for participants, allowing them to express genuine responses.

Document Findings: Document all research findings, including interview transcripts, usability test observations, survey responses, and so on.


**4. Data Analysis and Synthesis**

After the research, it is time to analyze and identify key themes and insights based on the data.

Identify Patterns: Look for recurring patterns in user feedback. What are the most frequent complaints, frustrations, and suggestions

Prioritize Issues: Rank the identified issues in order of impact on user engagement, and by the feasibility of addressing them.

Develop Personas: Create user personas to represent various segments of the user base.

**5. Solution Development and Implementation**

Based on the findings, formulate and implement solutions to address the identified issues.

Prioritize Fixes: Address the most crucial issues first. This may mean that the developer should fix bugs, improve usability, or revert to unwanted changes.

Iterate and Test: Continuously iterate on the solution in light of user feedback. Conduct A/B testing to determine what works best.

Communicate with Users: Communicate the progress of fixes and enhancements to the users.

Make them feel heard, and their feedback is valued.

**Conclusion**

Now, it is true that losing 30% of DAU is a big deal and, simultaneously, a great opportunity to learn and improve. A user-centered research approach helps us understand why this happened and what we could do about it. Let's not forget that, in everything, empathy, transparency, and gratitude are key. By listening to our users and catering to their needs, we can get back on track with the engagement of users and build a better app. This case study essentially identifies how user research and iteration need to be ongoing to ensure the success of any product.



 Balancing Business and User Goals . You’re the PM of an e-commerce platform. The business team wants to push flash sales aggressively, but users complain of a cluttered homepage and difficulty finding regular items. How would you balance both perspectives and move forward? Explain how you’d apply problem-solving principles, user empathy, and data-backed insights to navigate this challenge.


 **Introduction**

As a Product Manager, one of the most common challenges is balancing what the business wants with what the users actually need. In this scenario, the e-commerce business team wants to push flash sales more aggressively to increase short-term revenue. However, users are complaining that the homepage has become too cluttered and they are finding it difficult to locate regular items. My role as a PM is to find a middle ground that respects business objectives while protecting the overall user experience.

**Understanding the Problem**

The root of the issue is not the presence of flash sales themselves, but the way they dominate the homepage. When too many promotional banners appear at once, the interface becomes overwhelming. This creates friction for users who simply want to browse normally or quickly buy everyday essentials.

To better the understanding of the situation, I would look at:

* User feedback and ratings
* Search trends: For example, an acute rise in search usage
* Click patterns on the home page

* Drop-off and bounce rates

This helps me empathize with users and identify exactly where the confusion is coming from.

User Empathy

From a user’s perspective, a cluttered homepage creates unnecessary stress. They expect a clean, predictable layout where routine items are easily accessible.
Their needs are simple:

* An organized feeling home page

* Quick access to regular categories

* Relevant promotions, without being overwhelming

It means that if I put myself in the shoes of the users, the solution shall minimize visual clutter without removing business-driven promotions altogether.

**Data Backed Insights**

Before making any decisions on changes, I would look to data to validate either side:

* Flash sales may earn a high conversion rate, yet only within a certain section of the users.

* Heat maps may indicate that users tend to ignore most promotional sections placed at the top.

* Rising bounce rates could confirm that clutter is discouraging exploration.

* Search bar dependency suggests navigation difficulty.

This data ensures that decisions are not based on assumptions but are taken from real user behavior.

 **Solution Approach**

I would recommend a solution that will reorganize the homepage without removing flash sales to meet both business and user needs.

1. Customized Home Page Design

*   Instead of the same homepage for all, create variations that it gives a more promotion-heavy layout to users who frequently click on flash sales.

*  Regular customers, through this facility, view essential categories, “recently viewed”, and “frequently bought”.

This ensures relevance without overwhelming all users.

2. Cleaner Information Architecture



*   Move flash sales to a dedicated, visually-contained section.

*   Maintain the upper part of the home page as focused on navigation, search, and core categories.

Usability immediately benefits from a cleaner structure.

3. Combined Campaigns

Instead of several banners, display:

* A main flash sale card or

* An offer carousel that cycle through

This reduces clutter but retains visibility for promotion.

4. A/B Testing

Current flash-sale-heavy design Then, measure which one gives the best balance of Conversion and User satisfaction.

 Phase 1: Gather insights; define success metrics (CTR, bounce rate, homepage engagement).

 Phase 2: Design the new layout of the homepage; conduct usability tests with a small sample of users.

 Phase 3: Deploying gradually while performance monitoring and iterations go on.  

**Conclusion:**

Balancing business and user goals is not about choosing one over the other it is about finding a design solution that supports both. By using user empathy, data-based decision-making, and thoughtful UI restructuring, the platform can continue promoting flash sales without disrupting the daily shopping experience. This approach creates a healthier, long-term balance between revenue growth and user satisfaction.
