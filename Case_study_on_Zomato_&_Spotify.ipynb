**Case Study 1: Spotify – Podcasts & Exclusive Content Strategy**

**1. Key Pointers (Product Strategy)**

Spotify's product strategy was to add, beyond music, podcasts and exclusive audio content in order to:

Boost user interactions and session time.

Reduce reliance on music licensing.

Use owned content to create platform differentiation

The goal was to transition Spotify from simply a music app into an everyday audio destination.

**2. Clarifying Questions (Before Solutioning)**

To scope the problem correctly, I would first clarify:

Which region shall we be focusing on?

(Mature markets like the US vs high-growth markets like India)

For which age bracket would we be targeting?

18–24: discovery-led, experimentation

25–40: Podcasts, news and learning

**Which metric are we optimizing for?**

The session lasted for

Average time spent per user

Podcast follow and repeat listen rate

Current baseline performance:

How many music users also listen to podcasts?

Drop-off points within the podcast funnel

**3. Core Challenge Identified**

The most significant problem was how to discover podcasts.

Users knew that podcasts did exist on Spotify.

While discovery was intuitive, similar to music,

Podcasts were buried in a music-first interface.

New users did not have guidance on what to listen to and why. This resulted in poor adoption and weak habit formation.

4. Proposed Solutions (Lightweight Discovery Interventions

a. Tooltips **bold text**

Contextual tool tips to inform first-time users:

Following a Podcast derives from the

Resuming episodes

Personalized podcast recommendations

Shown progressively to avoid overload.

Goal: Reduce Learning Friction.

**b. Banners**

Home screen banners:

Trending Podcasts

Exclusive content

Personalized podcast recommendations

Triggered based upon user behavior (for instance, prolonged music use)

Goal: Increase visibility and trial.

**c. Push Notifications (PNs)**

Personalized notifications such as:

Episodes from the following series

Resume unfinished episodes

Recommendations based on listening history

Goal: Encourage repeat listens and behavioral habits.

**5. Expected Outcome**

 Improved Podcast Discovery  
Longer Session Duration
More time spent in the app   
Improved ROI from investments in exclusive content
